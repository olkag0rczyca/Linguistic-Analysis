## Linguistic analysis

As a student of Natural Language Processing, I have opted for linguistic analysis which focuses on *stemming* - **a process which produces variants of a base word**. \
The point of stemming is that by using fixed rules such as chopping off from either the end or beginning of the word-text we receive list of prefixes and suffixes that could be available in an infected word-text. \
Stemming has got a few algorithms, hovewer, for the purposes of this project, I will only introduce two of them:
1. **Porter Stemmer Algorithm**
2. **Regular Expression stemming Algorithm**




###  PORTER STEMMER ALGORITHM 

Thanks to PorterStemmer class we can convert the given text-word to the resultant stem which in turn gives a shorter word. \
Firstly, we have to import **nltk** library in order to then import PorterStemmer that has stemming rules.



```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

The variable *stemmer* holds Porterstemmer that we have just imported.


```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

Secondly, we have to tokenize the particular input. \
We use the **stem** function of the variable *stemmer* and on the each word in the list titled *words*.

```python
words = ["yours", "myself", "their", "theirs", "them", "they"]
for word in words:
    print(word, stemmer.stem(word))
```

### REGULAR EXPRESSION STEMMING ALGORITHM 

With the help of the Regexp Stemming algorithm, we can build our stemmer.\
NLTK library has a RegexpStemmer class that takes a single expression as input and removes suffix and prefix that matches the expression. 

## LEMMATISATION

Apart from stemming, it would be interesting to look at a similar process - *lemmatisation*.\
This process involves **separating out the base form of a word that convey the meaning**.\
The aim of this process is to try to find the lemma that is a root word.

Implementation of lemmatisation requires using NLTK since it provides WordNetLemmatizer class which is a slim cover wrapped around the WordnetCorpus.\
#### HOW TO LEMMATIZE THE WORD?
Firstly, we import nltk.stem to import WordNetLemmatizer. 
Secondly, we assign WordNetLemmatizer to variable called *lemmatizer* which has a function that lemmatize the word.



```python
from nltk.stem import WordNetLemmatizer
lemmatizer  = WordNetLemmatizer()
input_str = "apple mice feet cities"
input_str = nltk.word_tokenize(input_str)
for word in input_str:
    print(lemmatizer.lemmatize(word))
```

 After tokenization each word is getting lemmatized, each word holds the meaning.
